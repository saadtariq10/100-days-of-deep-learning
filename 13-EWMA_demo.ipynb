```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/DailyDelhiClimateTest.csv')
```

```python
df.head()
```

```python
df = df[['date','meantemp']]
```

```python
plt.scatter(df['date'],df['meantemp'], color='red')
```

```python
x1 = df['meantemp'].ewm(alpha=0.1).mean()
x2 = df['meantemp'].ewm(alpha=0.25).mean()
x3 = df['meantemp'].ewm(alpha=0.5).mean()
x4 = df['meantemp'].ewm(alpha=0.9).mean()
```

```python
df['x1'] = x1
df['x2'] = x2
df['x3'] = x3
df['x4'] = x4
```

```python
plt.figure(1)
plt.scatter(df['date'],df['x1'], color='red')
plt.plot(df['date'],df['x1'],color = 'black', label= 'alpha = 0.1')
plt.legend()
plt.show()

plt.figure(2)
plt.scatter(df['date'],df['x2'], color='green')
plt.plot(df['date'],df['x2'],color = 'black', label= 'alpha = 0.25)')
plt.legend()
plt.show()

plt.figure(3)
plt.scatter(df['date'],df['x3'], color='blue')
plt.plot(df['date'],df['x3'],color = 'black', label= 'alpha = 0.5')
plt.legend()
plt.show()

plt.figure(4)
plt.scatter(df['date'],df['x4'], color='yellow')
plt.plot(df['date'],df['x4'],color = 'black', label= 'alpha = 0.9')
plt.legend()
plt.show()
```

```python
# Plot all the figures in a single plot with different colors
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
plt.scatter(df['date'], df['meantemp'], color='gray', label='Original Data')
plt.plot(df['date'], df['x1'], color='red', label='alpha = 0.1')
plt.plot(df['date'], df['x2'], color='green', label='alpha = 0.25')
plt.plot(df['date'], df['x3'], color='blue', label='alpha = 0.5')
plt.plot(df['date'], df['x4'], color='yellow', label='alpha = 0.9')
plt.xlabel('Date')
plt.ylabel('Mean Temperature')
plt.title('Exponentially Weighted Moving Averages')
plt.legend()
plt.grid(True)
plt.show()
```
