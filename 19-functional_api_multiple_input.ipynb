```python
from keras.layers import *
from keras.models import Model
```

```python
inputA = Input(shape=(10,))
inputB = Input(shape=(10,))
```

```python
hidden1 = Dense(10, activation='relu')(inputA)
hidden2 = Dense(10, activation='relu')(hidden1)

hidden3 = Dense(10, activation='relu')(inputB)
hidden4 = Dense(10, activation='relu')(hidden3)
hidden5 = Dense(10, activation='relu')(hidden4)
```

```python
concat = concatenate([hidden2, hidden5])
```

```python
hidden6 = Dense(10, activation='relu')(concat)
output = Dense(1, activation='linear')(hidden6)
```

```python
model = Model(inputs=[inputA, inputB], outputs=output)
```

```python
from keras.utils import plot_model
plot_model(model)
```
